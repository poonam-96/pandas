```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
titnic_file_path = './train.csv'
data = pd.read_csv(titnic_file_path) 
print(data.columns)
data.shape
```

```python
del data['Cabin']
```

```python
data = data.dropna(axis=0, subset=['Embarked'])
```

```python
age_mean = int(data["Age"].mean())
```

```python
data['Age'].fillna(age_mean, inplace =True)
```

```python
del data['Name']
```

```python
data.sample()
```

```python
Y = data[['Survived']]
Y.head()
X = data.drop('Survived', axis = 1)
X.head()
```

```python
data['Embarked'].value_counts()
```

```python
def f(value):
    if value == 'S':
        return 0
    if value == 'C':
        return 1
    if value == 'Q':
        return 2
```

```python
data['Embarked'] = data['Embarked'].apply(lambda x: f(x))
```

```python
data
```

```python
def f2(y):
    if y == 'male':
        return 1
    if y == 'female':
        return 0
    
```

```python
data['Sex'] = data['Sex'].apply(lambda x: f2(x))
```

```python
Y = data[['Survived']]
Y.head()
X = data.drop('Survived', axis = 1)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,y_train)
predict_values = lr.predict(X_test)
```

```python
print(accuracy_score(y_test,predict_values))
print(confusion_matrix(y_test,predict_values))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predict_values))
```

```python
def Age(dataframe):
    dataframe.loc[dataframe['Age'] <= 30,'Age'] = 1
    dataframe.loc[(dataframe['Age'] > 30) & (dataframe['Age'] <= 40), 'Age'] = 2
    dataframe.loc[(dataframe['Age'] > 40) & (dataframe['Age'] <= 50), 'Age'] = 3
    dataframe.loc[(dataframe['Age'] > 50) & (dataframe['Age'] <= 60), 'Age'] = 4
    return dataframe

data3 = Age(data)
```

```python
data3
```

```python
data3 = pd.get_dummies(data3, columns=['PassengerId', 'Pclass', 'SibSp', 'Parch', 'Embarked', 'Age'])
Y = data3[['Survived']]
Y.head()
X = data3.drop('Survived', axis = 1)
X.head()

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,y_train)
predict_values = lr.predict(X_test)

print(accuracy_score(y_test,predict_values))
print(confusion_matrix(y_test,predict_values))

from sklearn.metrics import classification_report
print(classification_report(y_test,predict_values))
```
