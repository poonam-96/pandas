```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
melbourne_file_path = './train.csv'
data = pd.read_csv(melbourne_file_path) 
print(data.columns)
```

```python
data.head()
```

```python
price_data = data.SalePrice
price_data.head()
```

```python
a = ['LotArea', 'YrSold', 'SalePrice']
columns_data = data[a]
columns_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train  = ['LotArea', 'YrSold']
X = data[train]
X.head()
```

```python
test = ['SalePrice']
y = data[test]
y.head()
```

```python
X.head()
```

```python
tmp = X[['LotArea']]
```

```python
plt.figure(figsize=(15,5))
plt.plot(tmp)
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

X_train.shape, X_test.shape, y_train.shape, y_test.shape

```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)
```

```python
predict_values = model.predict(X_test)
predict_values
```

```python
# predict_values
```

```python
plt.figure(figsize=(15,5))
plt.plot(predict_values, label="predict_values")
plt.plot(y_test.values, label="y_test")
plt.legend()
```
